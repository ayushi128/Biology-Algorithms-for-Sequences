```python
#Brute Force Algorithm
from Bio.Seq import Seq

text = ("GATCACAGGTCTATCACCCTATTAACCACTCACGGGAGCTCTCCATGCATTTGGTATTTTCGTCTGGGGG"
        "GTATGCACGCGATAGCATTGCGAGACGCTGGAGCCGGAGCACCCTATGTCGCAGTATCTGTCTTTGATTC"
        "CTGCCTCATCCCATTATTTATCGCACCTACGTTCAATATTACAGGCGAACATACTTACTAAAGTGTGTTA")

dna = Seq(text)
prot = str(dna.translate().replace("*", ""))

print("Protein Sequence:",prot )

masstab = {
'A':89,'R':174,'N':132,'D':133,'C':121,'Q':146,'E':147,'G':75,
'H':155,'I':131,'L':131,'K':146,'M':149,'F':165,'P':115,'S':105,
'T':119,'W':204,'Y':181,'V':117
}

k = int(input("Enter the value of k: "))
def bruteforce(prot, k):
    max_weight = 0
    best = ""

    for i in range(len(prot)-k+1):
        peptide = prot[i:i+k]
        weight = sum(masstab[a] for a in peptide)

        if weight > max_weight:
            max_weight = weight
            best = peptide

    return best, max_weight

print("Brute Force:", bruteforce(prot, k))
```

```python
#Branch and Bound
from Bio.Seq import Seq

text = ("GATCACAGGTCTATCACCCTATTAACCACTCACGGGAGCTCTCCATGCATTTGGTATTTTCGTCTGGGGG"
        "GTATGCACGCGATAGCATTGCGAGACGCTGGAGCCGGAGCACCCTATGTCGCAGTATCTGTCTTTGATTC"
        "CTGCCTCATCCCATTATTTATCGCACCTACGTTCAATATTACAGGCGAACATACTTACTAAAGTGTGTTA")

dna = Seq(text)
prot = str(dna.translate().replace("*", ""))

print("Protein Sequence:",prot )

masstab = {
'A':89,'R':174,'N':132,'D':133,'C':121,'Q':146,'E':147,'G':75,
'H':155,'I':131,'L':131,'K':146,'M':149,'F':165,'P':115,'S':105,
'T':119,'W':204,'Y':181,'V':117
}

k = int(input("Enter the value of k: "))
def bnb(prot, k):
    max_weight = 0
    best = ""

    for i in range(len(prot)-k+1):
        weight = 0

        for j in range(k):
            weight += masstab[prot[i+j]]

            if weight < max_weight and j == k-1:
                break

        if weight > max_weight:
            max_weight = weight
            best = prot[i:i+k]

    return best, max_weight

print("Branch and Bound:", bnb(prot, k))
```

```python
#LeaderBoard
from Bio.Seq import Seq

text = ("GATCACAGGTCTATCACCCTATTAACCACTCACGGGAGCTCTCCATGCATTTGGTATTTTCGTCTGGGGG"
        "GTATGCACGCGATAGCATTGCGAGACGCTGGAGCCGGAGCACCCTATGTCGCAGTATCTGTCTTTGATTC"
        "CTGCCTCATCCCATTATTTATCGCACCTACGTTCAATATTACAGGCGAACATACTTACTAAAGTGTGTTA")

dna = Seq(text)
prot = str(dna.translate().replace("*", ""))

print("Protein Sequence:",prot )

masstab = {
'A':89,'R':174,'N':132,'D':133,'C':121,'Q':146,'E':147,'G':75,
'H':155,'I':131,'L':131,'K':146,'M':149,'F':165,'P':115,'S':105,
'T':119,'W':204,'Y':181,'V':117
}

k = int(input("Enter the value of k: "))
def leaderboard(prot, k, N):
    peptides = []

    for i in range(len(prot)-k+1):
        pept = prot[i:i+k]
        weight = sum(masstab[a] for a in pept)
        peptides.append((pept, weight))

    peptides.sort(key=lambda x: x[1], reverse=True)
    return peptides[:N]

print("Leaderboard Top 5:")
for p in leaderboard(prot, k, 5):
    print(p)
```
